2.2) Remove all the stopwords from the three speeches. Show the word count before and after the removal of stopwords. Show a sample sentence after the removal of stopwords.

```python
import string
stopwords=nltk.corpus.stopwords.words('english')+list(string.punctuation)+['--','."','...']
```

```python
print(stopwords)
```

```python
all_words_R = list(inaugural.words(inaugural.fileids()[38]))
all_words_K = list(inaugural.words(inaugural.fileids()[43]))
all_words_N = list(inaugural.words(inaugural.fileids()[46]))
```

```python
def clean(x):
    all_words = []
    for w in x:
        words=w.lower()
        if words not in stopwords:
            all_words.append(words)
    return all_words
```

```python
all_words_R_c = []
all_words_K_c = []
all_words_N_c = []
```

```python
all_words_R_c = clean(all_words_R)
all_words_K_c = clean(all_words_K)
all_words_N_c = clean(all_words_N)
```

```python
print("Word Count Before Removal of stopwords in Roosevelt speech:",len(inaugural.words(inaugural.fileids()[38])))
print("Word Count Before Removal of stopwords in Kennedy speech:",len(inaugural.words(inaugural.fileids()[43])))
print("Word Count Before Removal of stopwords in Nixon speech:",len(inaugural.words(inaugural.fileids()[46])))
```

```python
print("Word Count After Removal of stopwords in Roosevelt speech:",len(all_words_R_c))
print("Word Count After Removal of stopwords in Kennedy speech:",len(all_words_K_c))
print("Word Count After Removal of stopwords in Nixon speech:",len(all_words_N_c))
```

2.3) Which word occurs the most number of times in his inaugural address for each president? Mention the top three words.

```python
all_words_freq_R = nltk.FreqDist(all_words_R_c)
all_words_freq_K = nltk.FreqDist(all_words_K_c)
all_words_freq_N = nltk.FreqDist(all_words_N_c)
```

```python
print("The most common words and their frequency in Roosevelt speech:",all_words_freq_R.most_common(3))
```

```python
print("The most common words and their frequency in Kennedy speech:",all_words_freq_K.most_common(3))
```

```python
print("The most common words and their frequency in Nixon speech:",all_words_freq_N.most_common(3))
```

2.4) Plot the word cloud of each of the three speeches.

```python
from wordcloud import WordCloud
wc_R = ' '.join(all_words_R_c)
wc = WordCloud(width = 3000, 
               height =3000,
               background_color = 'blue',
               min_font_size=10,
               random_state = 100).generate(wc_R)
plt.figure(figsize=(10,10))
plt.imshow(wc)
plt.axis('off')
plt.title('Word Cloud for Roosevelt speech')
plt.show()
```

```python
wc_K = ' '.join(all_words_K_c)
wc = WordCloud(width = 3000, 
               height =3000,
               background_color = 'blue',
               min_font_size=10,
               random_state = 100).generate(wc_K)
plt.figure(figsize=(10,10))
plt.imshow(wc)
plt.axis('off')
plt.title('Word Cloud for Kennedy speech')
plt.show()
```

```python
wc_N = ' '.join(all_words_N_c)
wc = WordCloud(width = 3000, 
               height =3000,
               background_color = 'blue',
               min_font_size=10,
               random_state = 100).generate(wc_N)
plt.figure(figsize=(10,10))
plt.imshow(wc)
plt.axis('off')
plt.title('Word Cloud for Nixon speech')
plt.show()
```
